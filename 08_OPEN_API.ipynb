````python
# 0 ~ 999
# 1000 ~ 1999
# 2000 ~ 2999
# 3000 ~ 3171
base_url = "http://openapi.seoul.go.kr:8088"

# 요청 파라미터
KEY = os.getenv("SEOUL_DATA_API_KEY")
N_TYPE = "json"
SERVICE = "tbCycleStationInfo"

# 0. 빈 리스트를 만든다 (total_data)
# 1. start_index가 0, 1000, 2000, 3000
# 2. end를 만든다. end = start + 1000 -1
# 3. url을 만든다
# 4. 요청
# 5. 데이터 리스트를 파싱
# 6. total_data에 data_list를 합친다 (extend)
# 7. total_data를 pd.DataFrame()을 이용해서 데이터프레임을 만든다
total_data = []
for start in range(0, 3001, 1000) :
    end = start + 1000 -1

    url = f"{base_url}/{KEY}/{N_TYPE}/{SERVICE}/{start}/{end}"
    # print(url)
    response = requests.get(url)
    response_json = response.json()
    data_list = response_json["stationInfo"]["row"]
    print(len(data_list))
    total_data.extend(data_list) 
# print(total_data)
# total_data
data = pd.DataFrame(total_data)
data
````

## 2) requests 활용

````python
# 네이버 검색 API 예제 - 블로그 검색
import os
import sys
import urllib.request
from dotenv import load_dotenv
load_dotenv()

# 1. 정보 입력
client_id = os.getenv("NAVER_CLIENT_ID")
client_secret = os.getenv("NAVER_CLIENT_SECRET")

# 2. 키워드 입력
keyword = "인공지능"

# 3. 키워드 인코딩하기
encText = urllib.parse.quote(keyword)
# print(keyword)
# print(encText)

# 4. URL 만들기
base_url = "https://openapi.naver.com/v1/search/news.json"
query = encText

url = f"{base_url}?query={query}"

# 5. 데이터 요청
headers = {
    "X-Naver-Client-Id" : client_id,
    "X-Naver-Client-Secret" : client_secret
}

response = requests.get(url, headers=headers)
if response.status_code == 200 :
    print(response.json())
else:
    print("Error")
````

## 3) requests 활용 2

````python
# 네이버 검색 API 예제 - 블로그 검색
import os
import sys
import urllib.request
from dotenv import load_dotenv
load_dotenv()

# 1. 정보 입력
client_id = os.getenv("NAVER_CLIENT_ID")
client_secret = os.getenv("NAVER_CLIENT_SECRET")

# 2. 키워드 입력
keyword = "인공지능"

# 3. 파라미터 만들기
params = {
    "query" : keyword,
    "display" : 33,
    "sort" : "date"
}

# 4. URL 만들기
base_url = "https://openapi.naver.com/v1/search/news.json"

url = f"{base_url}?query={query}"

# 5. 데이터 요청
headers = {
    "X-Naver-Client-Id" : client_id,
    "X-Naver-Client-Secret" : client_secret
}

response = requests.get(url, headers=headers, params=params)
if response.status_code == 200 :
    response_json = response.json()
    print(response.json())
else:
    print("Error")
````

````python
# 유튜브 댓글을 수집하는 함수
from googleapiclient.discovery import build 
import pandas as pd
import os 
from dotenv import load_dotenv 
load_dotenv()

api_key = os.getenv("YOUTUBE_API_KEY")
video_id = "7qXFLRRMbKw"

youtube = build("youtube", "v3", developerKey=api_key)

comments = []
next_page_token = None 

while True:
    response = youtube.commentThreads().list(
        part="snippet",
        videoId=video_id,
        maxResults=100,
        pageToken=next_page_token,
        textFormat="plainText"
    ).execute()

    for item in response["items"]:
        comment = item["snippet"]["topLevelComment"]["snippet"]
        comments.append(
            {
                "author": comment["authorDisplayName"],
                "text": comment["textDisplay"],
                "likeCount": comment["likeCount"],
                "publishedAt": comment["publishedAt"]
            }
        )

    next_page_token = response.get("nextPageToken")
    if not next_page_token:
        break 

data = pd.DataFrame(comments)
data.head()
        
````
