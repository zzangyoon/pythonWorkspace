## 1) 범주형 데이터

````python
import seaborn as sns
import matplotlib.pyplot as plt
import koreanize_matplotlib
````

````python
plt.figure(figsize=(3,3))
sns.countplot(data=data, x="성별")  # 막대그래프 그릴때
plt.title("성별에 대한 막대그래프")
plt.show()
````

## 2) 순서형 데이터

````python
# 만족도에 대한 막대 그래프 그리기
plt.figure(figsize=(3,3))
sns.countplot(data=data, x="만족도", order=["하", "중", "상"])
plt.title("만족도에 대한 막대그래프")
plt.show()
````

````python
# 성별에 따른 만족도 그래프
plt.figure(figsize=(3,3))
sns.countplot(data=data, x="만족도", order=["하", "중", "상"], hue="성별")
plt.title("성별에 따른 만족도 그래프")
plt.show()
````

## 3) 연속형 데이터

````python
# 연속형, 이산형 같은 데이터들을 연속으로 나타낸 그래프
# "월 수입(만원)"열에 대한 히스토그램
plt.figure(figsize=(3,3))
plt.hist(data["월 수입(만원)"], bins = 20, edgecolor = "black")
plt.title("월 수입(만원) 히스토그램")
plt.show()
````

````python
plt.figure(figsize=(3, 3))
plt.boxplot(x=data["월 수입(만원)"])
plt.title("월 수입(만원)에 대한 Box Plot")
plt.xlabel("월 수입(만원)")
plt.show()
````

````python
# botxplot 을 sns 로 그리기
plt.figure(figsize=(3, 3))
# sns.boxplot(data=data, y = "월 수입(만원)")
sns.boxplot(data=data, x = "월 수입(만원)")
plt.xlabel("월 수입(만원)")
plt.show()
````

````python
# 성별에 따른 월 수입 상자도표 그리기
plt.figure(figsize=(3, 3))
sns.boxplot(data=data, x="성별", y = "월 수입(만원)")
plt.title("성별 월 수입(만원) Box Plot")
plt.show()
````

## 4) 이산형 데이터

````python
import matplotlib.pyplot as plt
plt.subplots(figsize=(3,3))
````

````python
# subplots 의 결과물이 튜플로 나온다 -> 변수에 담는다
# -> 변수 하나에 담게 되면 a =(<Figure size 300x300 with 1 Axes>, <Axes: >) 되어서 a = ( - , - ) 로 나오게 되는 것이고,
# -> fig, ax = ( , ) 이렇게 담게 된다면 fig, ax = (<Figure size 300x300 with 1 Axes> , <Axes: >) 이런식으로 담게 되는것이다
````

````python
# 방문 페이지 수
## 구글링 : ax를 이용한 시각화 도전
fig, ax = plt.subplots(figsize=(3,3))
sns.(datacountplot=data, x="방문 페이지 수", ax=ax)
##### 타이틀 작성하는 방법 #####
ax.set_title("방문 페이지 수에 대한 막대그래프")
plt.show()

# plt, ax 옵션을 표현하는 방법들이 다름
````

### 날짜 열을 datetime으로 바꾸는 방법 ||

````python
data = pd.read_csv("data/sample_data.csv")
data.info()
````

````python
data["날짜"] = pd.to_datetime(data["날짜"])
data.info()
````

````python
fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(data["날짜"], data["평균 체류 시간(분)"], marker="o")   # marker="0" 점찍어줌
ax.set_title("날짜별 평균 체류 시간 변화 그래프")
ax.set_xlabel("날짜")
ax.set_ylabel("평균 체류시간(분)")
plt.show()
````

# subplots

````python
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
plt.subplots(figsize=(3,3))
````

````python
plt.subplots(1, 3, figsize=(10,3))
# 반복문으로 똑같은 그래프를 그릴때 (다른데이터) 주로 사용
````

````python
fig, ax = plt.subplots(1, 3, figsize=(10,3))
# 첫번째 그래프: 자녀 수에 대한 막대 그래프
sns.countplot(data=data, x="자녀 수", ax=ax[0])
ax[0].set_title("자녀 수에 대한 막대그래프")
# 두번째 그래프: 평균 체류 시간(분)에 대한 히스토그램
ax[1].hist(data["평균 체류 시간(분)"], edgecolor="black")
ax[1].set_title("평균 체류 시간(분)에 대한 히스토그램")
# 세번째 그래프: 월 수입(만원)에 대한 상자그림
# ax[2].boxplot(x=data["월 수입(만원)"], vert=False)
sns.boxplot(y=data["월 수입(만원)"], ax=ax[2])
ax[2].set_title("월 수입(만원)에 대한 Box Plot")
ax[2].set_xlabel("월 수입(만원)")
plt.show()
````
